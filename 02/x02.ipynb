```python
from numpy import genfromtxt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

# Another way
dataset = genfromtxt('x02.csv',delimiter=',', skip_header=0, dtype=(int, int, int, int))

dataset = pd.read_csv('x02.csv', names=["gender", "age", "head_size", "brain_weight"])

print(type(dataset))

x = dataset['brain_weight'].values
y = dataset['head_size'].values

# Prend un polynome de degré 1 et trouve les coef pour faire la droite
polynomCoef = np.polyfit(x,y,deg=1)
# Avec ces coef, on va se faire une fonction
h = np.poly1d(polynomCoef)
# y^ = h(x)
xUnseen = np.linspace(np.min(x), np.max(x))
yHat = h(xUnseen)

plt.scatter(x,y)
plt.plot(xUnseen,yHat, color='red', label='deg 1')

polynomCoef = np.polyfit(x,y,deg=2)
h = np.poly1d(polynomCoef)
yHat = h(xUnseen)
plt.plot(xUnseen,yHat, color='yellow', label='deg 2')

plt.xlabel('brain weight')
plt.ylabel('head size')
plt.legend()
plt.show()

```
